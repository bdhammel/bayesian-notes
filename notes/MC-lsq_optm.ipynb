```python
def LossFn():
    x = np.linspace(-5, 5)
    y = x * m 
    
    def _loss_fn(w):
        return np.mean((y - x*w)**2)
                       
    return _loss_fn
                   
loss_fn = LossFn()  
eta = 9e-2

_w = np.random.uniform(low=-3, high=3)
_loss = loss_fn(_w)
dLdw = 1e-3

ws = []
losses = []

for _ in range(25):
    w = _w - eta * dLdw
    loss = loss_fn(w)
    dLdw = (loss - _loss) / (w - _w)
    _loss = loss
    _w = w
    ws.append(w)
    losses.append(loss)
```

```python
plt.plot(ws, losses)
```

```python
w = np.random.uniform(low=-3, high=3)
loss = loss_fn(_w)
dLdw = 1e-3

ws = []
losses = []

for _ in range(1000):
    _w = np.random.normal(loc=w, scale=1e-1)
    _loss = loss_fn(_w)
    
    rho = min(1, loss/_loss)
    
    if rho > np.random.uniform():
        loss = _loss
        w = _w
        
    ws.append(w)
    losses.append(loss)
```
